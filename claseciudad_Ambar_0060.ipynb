```python
class Ciudad:
    def __init__(self, nombre, poblacion, pais):
        self.nombre = nombre
        self.poblacion = poblacion
        self.pais = pais

    def leerdatos(self):
        self.nombre = input("Ingrese el nombre de la ciudad: ")
        self.poblacion = input("Ingrese la población: ")
        self.pais = input("Ingrese el país: ")
        return self.nombre, self.poblacion, self.pais

    def mostrardatos(self):
        print("Nombre: ", self.nombre)
        print("Población: ", self.poblacion)
        print("País: ", self.pais)
        return self.nombre, self.poblacion, self.pais

# Fin de clase Ciudad

print("--- Registro de Ciudad 1 ---")
# Creando objeto ciudad1 con valores iniciales vacíos
ciudad1 = Ciudad("", "", "")
ciudad1.leerdatos()
ciudad1.mostrardatos()

print("-*-*-*-*-*-*-*-*-*-*-*")

print("--- Registro de Ciudad 2 ---")
# Creando objeto ciudad2 con valores iniciales vacíos
ciudad2 = Ciudad("", "", "")
ciudad2.leerdatos()
ciudad2.mostrardatos()

print("-*-*-*-*-*-*-*-*-*-*-*")
print("Software desarrollado por Ambar Fierro NL = 19")
```
